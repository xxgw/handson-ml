```python
from tensorflow.examples.tutorials.mnist import input_data
import tensorflow as tf
```

```python
mnist = input_data.read_data_sets("../datasets/MNIST_data")
```

```python
X_train = mnist.train.images
X_test = mnist.test.images
y_train = mnist.train.labels.astype("int")
y_test = mnist.test.labels.astype("int")
```

```python
config = tf.contrib.learn.RunConfig(tf_random_seed = 42)
feature_columns = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units=[300, 100], n_classes=10, feature_columns=feature_columns)
dnn_clf = tf.contrib.learn.SKCompat(dnn_clf) # if TensorFlow >= 1.1
dnn_clf.fit(x=X_train, y=y_train, batch_size=50, steps=40000)
```

```python
from sklearn.metrics import accuracy_score

y_pred = dnn_clf.predict(X_test)
accuracy_score(y_test, y_pred['classes'])
```

## 用Tensorflow处理质量数据

```python
import os
import numpy as np
import pandas as pd

QUALITY_PATH = os.path.join("..\datasets", "quality")
cvs_path = os.path.join(QUALITY_PATH, "quality.csv")
quality = pd.read_csv(cvs_path)
```

```python
# 训练集和测试集数据
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for train_index, test_index in split.split(quality, quality["Quality"]):
    strat_train_set = quality.loc[train_index]
    strat_test_set = quality.loc[test_index]

X_train = strat_train_set.drop(["PRD", "Quality"], axis = 1)
y_train = strat_train_set["PRD"].copy()
```

```python
import tensorflow as tf
config = tf.contrib.learn.RunConfig(tf_random_seed = 42)
feature_columns = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units=[300, 100], n_classes=10, feature_columns=feature_columns)
dnn_clf = tf.contrib.learn.SKCompat(dnn_clf) # if TensorFlow >= 1.1
dnn_clf.fit(x=X_train, y=y_train, batch_size=5, steps=5)
```
